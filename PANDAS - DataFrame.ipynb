# DataFrame 

- DataFrame is a two dimensional data structure, like a two dimensional array or a table with rows and columns As
- Simply a DataFrame is a collection of series, with at least one series.

```python
import pandas as pd
```

### Based on the name Kelahiran Bayi_Jakarta_2020,
I assume it is a dataset or file related to the births of babies in Jakarta, Indonesia, for the year 2020. 

## Load a Data CSV

```python
df = pd.read_csv('Kelahiran Bayi_Jakarta_2020.csv') # Load A Data CSV
```

## Head
The head() function is used to get the first rows.head by default top five

```python
df.head() 
```

## Info
Info method prints information about the DataFrame

```python
df.info() 
```

## Notnull
Notnull used to finds values that are NOT NULL

```python
# The following code calls to see the number of non null data

df.notnull().sum()
```

## Isnull
 Isnull used to finds NULL values in DataFrame

```python
# The following code calls to see the number of nulls in the data

df.isnull().sum()
```

## Tail
The tail() function is used to get the last five lines by default.

```python
#The following code is to see the order of the bottom data

df.tail() 
```

## Describe 
Returns a description summary for each column in the DataFrame,describe contains numeric data from managed datasets.

```python
#The following code displays information from columns in the form of numbers

df.describe()
```

## Mean 
Mean (The average value) = Return the mean of the values in the specified axis.

```python
#for example displays the average of the Amount column

df['jumlah'].mean() 
```

## Median
Median (The mid point value) = Return the median of the values in the specified axis.

```python
#for example displays the median of the Number column

df['jumlah'].median() 
```

## Mode 
Mode (The most common value) = Returns the mode of the values in the specified axis

```python
#for example displays the mode of the Number column

df['jumlah'].mode()[0] 
```

## Shape 
Shape is the number of rows and columns on the index of the DataFrame.

```python
# see the number of rows and number of columns

df.shape 
```

## Column 
Column returns the label of each column in the DataFrame

```python
# see the columns

df.columns
```

```python
#how to call a specific column

df[['kecamatan','kelurahan', 'jenis_kelamin', 'jumlah']] 
```

```python
#call column number whose value is 2

df[df['jumlah']==2]
```

```python
#call month columns that are more than 5

df[df['bulan']>5]
```

```python
#call the month column that is less than equal to 5

df[~(df['bulan']<5)]
```

```python
#call the month column more than 5 and the gender is female

df[(df['bulan']>5) & (df['jenis_kelamin'] =='Perempuan')] 
```

```python
#call the month column more than 5 or female gender

df[(df['bulan']>5) | (df['jenis_kelamin'] =='Perempuan')]  
```

```python
# - call the month column whose value is between 5 and 10

df[((df['bulan']>5) & (df['bulan']<10))] 
```

```python
# Call the City, Gender, with Count Beetween 20 column 
```

```python
df[['kota_kabupaten','kelurahan', 'jenis_kelamin', 'jumlah']][df['jumlah']>20 | (df['jenis_kelamin'] =='Laki-Laki')]
```
